There's a clear unbalance in the data, 9555 sample per class is fair enough for a finetuning task

```python
N = 9555

balanced_df = pd.concat([
    data[data["Target"] == 1],
    data[data["Target"] == 0].sample(n=N, random_state=42)
], ignore_index=True)

balanced_df.head()
```

```python
downsamplde_data = balanced_df.sample(frac=0.001, ignore_index=True)
downsamplde_data.sort_values(by="patientId", inplace=True)
downsamplde_data.shape
downsamplde_data.head()
```

```python
downsamplde_data.to_csv("downsampled_data.csv", index=False)
```
